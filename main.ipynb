```python
# Linear Regression dengan L1 (Lasso) dan L2 (Ridge) Regularization
# Implementasi GD dan SGD untuk Tugas S3

"""
Notebook ini mengimplementasikan regresi linier dengan regularisasi L1 (Lasso) dan L2 (Ridge)
menggunakan Gradient Descent (GD) dan Stochastic Gradient Descent (SGD).
Setiap kombinasi hyperparameter diuji, dievaluasi dengan MSE dan R²,
serta visualisasi konvergensi loss dan perbandingan kecepatan konvergensi.
"""

# 1. Persiapan Data dan Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import time
from itertools import product
from tqdm import tqdm

# Set styling untuk visualisasi
plt.style.use('seaborn-v0_8-whitegrid')
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

# 1.1 Load dataset (Medical Cost Personal Dataset)
# Anda dapat mengubah path file sesuai lokasi dataset
df = pd.read_csv('Medical Cost Personal.csv')
print("Dataset Shape:", df.shape)
print("\nPeek at data:")
print(df.head())

# 1.2 Info dan ringkasan statistik dataset
print("\nInformasi Dataset:")
print(df.info())
print("\nRingkasan Statistik:")
print(df.describe())

# 1.3 Cek missing values
print("\nJumlah Missing Values per Kolom:")
print(df.isnull().sum())

# 1.4 Drop missing values (jika ada)
df.dropna(inplace=True)

# 1.5 One-hot encoding untuk variabel kategorikal
print("\nKolom sebelum encoding:")
print(df.columns.tolist())
df_encoded = pd.get_dummies(df, drop_first=True).astype(float)
print("\nKolom setelah encoding:")
print(df_encoded.columns.tolist())

# 1.6 Pisahkan fitur dan target
y = df_encoded.pop('charges').values
X = df_encoded.values

# 1.7 Standardisasi fitur
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 1.8 Split data menjadi train-test (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# 1.9 Tambahkan bias (intercept)
def add_bias(X):
    return np.hstack([np.ones((X.shape[0], 1)), X])

X_train_bias = add_bias(X_train)
X_test_bias = add_bias(X_test)

print(f"\nDimensi Data Training setelah preprocessing: {X_train_bias.shape}")
print(f"Dimensi Data Testing setelah preprocessing: {X_test_bias.shape}")

# 2. Exploratory Data Analysis (EDA)
print("\n---- Exploratory Data Analysis ----")

# 2.1 Distribusi target variable (charges)
plt.figure(figsize=(10, 6))
sns.histplot(df['charges'], bins=30, kde=True)
plt.title('Distribusi Biaya (Charges)', fontsize=15)
plt.xlabel('Charges', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)
plt.tight_layout()
plt.savefig('charges_distribution.png')
plt.show()

# 2.2 Matriks korelasi fitur
plt.figure(figsize=(12, 10))
correlation_matrix = df_encoded.corr()
mask = np.triu(correlation_matrix)
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm',
            mask=mask, linewidths=0.5)
plt.title('Matriks Korelasi Fitur', fontsize=15)
plt.tight_layout()
plt.savefig('correlation_matrix.png')
plt.show()

# 2.3 Scatter plot antara fitur numerik dan target
numeric_cols = ['age', 'bmi', 'children']
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, col in enumerate(numeric_cols):
    sns.scatterplot(data=df, x=col, y='charges', ax=axes[i])
    axes[i].set_title(f'Charges vs {col.capitalize()}', fontsize=13)
    axes[i].set_xlabel(col.capitalize(), fontsize=11)
    axes[i].set_ylabel('Charges', fontsize=11)

plt.tight_layout()
plt.savefig('feature_target_relationships.png')
plt.show()

# 2.4 Box plot untuk fitur kategorikal
categorical_cols = ['sex', 'smoker', 'region']
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, col in enumerate(categorical_cols):
    sns.boxplot(data=df, x=col, y='charges', ax=axes[i])
    axes[i].set_title(f'Charges by {col.capitalize()}', fontsize=13)
    axes[i].set_xlabel(col.capitalize(), fontsize=11)
    axes[i].set_ylabel('Charges', fontsize=11)

plt.tight_layout()
plt.savefig('categorical_features_boxplots.png')
plt.show()

# 3. Definisi Model Regresi Linear dengan Regularisasi
class LinearRegressionGD:
    """
    Implementasi Linear Regression dengan Gradient Descent dan regularisasi.

    Parameters:
    -----------
    lr : float
        Learning rate untuk gradient descent
    epochs : int
        Jumlah epochs untuk training
    penalty : {None, 'l1', 'l2'}, default=None
        Tipe regularisasi: None=tanpa regularisasi, 'l1'=Lasso, 'l2'=Ridge
    lam : float, default=0.0
        Parameter regularisasi lambda
    batch_size : int atau None, default=None
        Ukuran batch untuk SGD. None = full batch GD
    """

    def __init__(self, lr=0.01, epochs=1000, penalty=None, lam=0.0, batch_size=None):
        self.lr = lr
        self.epochs = epochs
        self.penalty = penalty  # None, 'l1', 'l2'
        self.lam = lam
        self.batch_size = batch_size
        self.theta = None
        self.history = []  # untuk menyimpan loss per epoch
        self.theta_history = []  # untuk menyimpan koefisien per epoch

    def _compute_gradient(self, X_batch, y_batch):
        """Menghitung gradient dari loss function dengan/tanpa regularisasi"""
        m = X_batch.shape[0]
        preds = X_batch.dot(self.theta)
        error = preds - y_batch

        # Gradient dari MSE
        grad = X_batch.T.dot(error) / m

        # Tambahkan gradient dari penalty regularisasi
        if self.penalty == 'l2':  # Ridge: L2 penalty
            # Tidak menerapkan regularisasi pada bias (theta[0])
            grad[1:] += (self.lam / m) * self.theta[1:]
        elif self.penalty == 'l1':  # Lasso: L1 penalty
            # Tidak menerapkan regularisasi pada bias (theta[0])
            grad[1:] += (self.lam / m) * np.sign(self.theta[1:])

        return grad

    def _compute_loss(self, X, y):
        """Menghitung loss (MSE + regularisasi)"""
        m = X.shape[0]
        preds = X.dot(self.theta)
        mse = np.mean((preds - y) ** 2)

        # Tambahkan penalty regularisasi ke loss
        if self.penalty == 'l2':  # Ridge: L2 penalty
            mse += (self.lam / (2 * m)) * np.sum(self.theta[1:] ** 2)
        elif self.penalty == 'l1':  # Lasso: L1 penalty
            mse += (self.lam / m) * np.sum(np.abs(self.theta[1:]))

        return mse

    def fit(self, X, y):
        """Fit model dengan Gradient Descent atau Stochastic Gradient Descent"""
        n_samples, n_features = X.shape
        self.theta = np.zeros(n_features)

        # Tentukan batch size
        bs = self.batch_size if self.batch_size else n_samples

        # Training loop
        for epoch in range(self.epochs):
            # Shuffle data untuk SGD
            if bs < n_samples:
                indices = np.random.permutation(n_samples)
                X_shuffled, y_shuffled = X[indices], y[indices]
            else:
                X_shuffled, y_shuffled = X, y

            # Mini-batch updates
            for start_idx in range(0, n_samples, bs):
                end_idx = min(start_idx + bs, n_samples)
                X_batch = X_shuffled[start_idx:end_idx]
                y_batch = y_shuffled[start_idx:end_idx]

                # Hitung gradient dan update parameters
                grad = self._compute_gradient(X_batch, y_batch)
                self.theta -= self.lr * grad

                # Early stopping jika weights tidak finite
                if not np.isfinite(self.theta).all():
                    print(f"Warning: Non-finite weights detected at epoch {epoch}. Training stopped.")
                    return self

            # Hitung dan simpan loss tiap epoch
            loss = self._compute_loss(X, y)
            self.history.append(loss)

            # Simpan theta history untuk analisis
            self.theta_history.append(self.theta.copy())

            # Early stopping jika loss tidak finite atau terlalu kecil (konvergen)
            if not np.isfinite(loss):
                print(f"Warning: Non-finite loss detected at epoch {epoch}. Training stopped.")
                return self

            # Print progress setiap 100 epoch atau di epoch terakhir
            if epoch % 100 == 0 or epoch == self.epochs - 1:
                print(f"Epoch {epoch}/{self.epochs}, Loss: {loss:.6f}")

        return self

    def predict(self, X):
        """Prediksi dengan model yang sudah dilatih"""
        return X.dot(self.theta)

    def get_params(self):
        """Return model parameters"""
        return {'intercept': self.theta[0], 'coefficients': self.theta[1:]}

# 4. Pengaturan Hyperparameter dan Tuning
print("\n---- Hyperparameter Tuning ----")

# Definisi hyperparameter untuk di-tuning
hyperparams = {
    'lr': [0.01, 0.1, 0.5, 1.0],
    'epochs': [100, 500, 1000],
    'batch_size': [1, 16, 32, None],
    'lam': [0.001, 0.01, 0.1, 1.0]
}

# Buat semua kombinasi hyperparameter
combinations = []
for penalty in [None, 'l1', 'l2']:
    # Hanya menggunakan lambda > 0 jika ada regularisasi
    lam_list = hyperparams['lam'] if penalty else [0.0]

    for lr, epochs, bs, lam in product(
        hyperparams['lr'],
        hyperparams['epochs'],
        hyperparams['batch_size'],
        lam_list
    ):
        combinations.append({
            'penalty': penalty,
            'lr': lr,
            'epochs': epochs,
            'batch_size': bs,
            'lam': lam
        })

print(f"Total kombinasi hyperparameter: {len(combinations)}")

# Proses tuning dengan semua kombinasi
results = []
for params in tqdm(combinations, desc='Tuning Models'):
    # Set parameter untuk model ini
    model_params = params.copy()

    try:
        # Catat waktu mulai
        t0 = time.time()

        # Inisialisasi dan latih model
        model = LinearRegressionGD(**model_params)
        model.fit(X_train_bias, y_train)

        # Prediksi pada data test
        y_pred = model.predict(X_test_bias)

        # Evaluasi model
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)

        # Catat waktu selesai
        t1 = time.time()

        # Simpan hasil
        results.append({
            'Penalty': params['penalty'] if params['penalty'] else 'None',
            'Learning Rate': params['lr'],
            'Epochs': params['epochs'],
            'Batch Size': params['batch_size'] if params['batch_size'] else 'Full',
            'Lambda': params['lam'],
            'MSE': mse,
            'R2': r2,
            'Training Time (s)': t1 - t0
        })
    except Exception as e:
        print(f"Error with params {params}: {e}")
        results.append({
            'Penalty': params['penalty'] if params['penalty'] else 'None',
            'Learning Rate': params['lr'],
            'Epochs': params['epochs'],
            'Batch Size': params['batch_size'] if params['batch_size'] else 'Full',
            'Lambda': params['lam'],
            'MSE': np.nan,
            'R2': np.nan,
            'Training Time (s)': np.nan
        })

# 5. Evaluasi Model
print("\n---- Evaluasi Model ----")

# Buat DataFrame hasil dan hapus baris dengan nilai NaN
results_df = pd.DataFrame(results).dropna()

# Urutkan berdasarkan MSE (dari terendah ke tertinggi)
sorted_results = results_df.sort_values('MSE').reset_index(drop=True)

# Cetak 10 model terbaik
print("\nTop 10 Model Berdasarkan MSE Terendah:")
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: '%.4f' % x)
print(sorted_results.head(10))

# 5.1 Pilih model terbaik berdasarkan MSE untuk analisis lebih lanjut
best_model_params = sorted_results.iloc[0].to_dict()
print("\nParameter Model Terbaik:")
for key, value in {k: best_model_params[k] for k in
                  ['Penalty', 'Learning Rate', 'Epochs', 'Batch Size', 'Lambda']}.items():
    print(f"{key}: {value}")

# 5.2 Visualisasi perbandingan performa model berdasarkan tipe regularisasi
plt.figure(figsize=(12, 6))
sns.boxplot(x='Penalty', y='MSE', data=results_df)
plt.title('Perbandingan MSE Berdasarkan Tipe Regularisasi', fontsize=15)
plt.xlabel('Tipe Regularisasi', fontsize=12)
plt.ylabel('Mean Squared Error (MSE)', fontsize=12)
plt.tight_layout()
plt.savefig('regularization_comparison_mse.png')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='Penalty', y='R2', data=results_df)
plt.title('Perbandingan R² Berdasarkan Tipe Regularisasi', fontsize=15)
plt.xlabel('Tipe Regularisasi', fontsize=12)
plt.ylabel('R² Score', fontsize=12)
plt.tight_layout()
plt.savefig('regularization_comparison_r2.png')
plt.show()

# 5.3 Visualisasi perbandingan performa model berdasarkan batch size
plt.figure(figsize=(12, 6))
sns.boxplot(x='Batch Size', y='MSE', data=results_df)
plt.title('Perbandingan MSE Berdasarkan Ukuran Batch', fontsize=15)
plt.xlabel('Ukuran Batch', fontsize=12)
plt.ylabel('Mean Squared Error (MSE)', fontsize=12)
plt.tight_layout()
plt.savefig('batch_size_comparison_mse.png')
plt.show()

# 5.4 Visualisasi pengaruh learning rate
plt.figure(figsize=(12, 6))
sns.boxplot(x='Learning Rate', y='MSE', data=results_df)
plt.title('Perbandingan MSE Berdasarkan Learning Rate', fontsize=15)
plt.xlabel('Learning Rate', fontsize=12)
plt.ylabel('Mean Squared Error (MSE)', fontsize=12)
plt.tight_layout()
plt.savefig('learning_rate_comparison_mse.png')
plt.show()

# 6. Perbandingan GD vs SGD
print("\n---- Perbandingan GD vs SGD ----")

# 6.1 Latih model GD (full-batch) dan SGD dengan parameter terbaik
penalty_best = 'l2' if best_model_params['Penalty'] == 'l2' else ('l1' if best_model_params['Penalty'] == 'l1' else None)
lr_best = best_model_params['Learning Rate']
lam_best = best_model_params['Lambda']
epochs = 500  # menggunakan 500 epochs untuk visualisasi yang jelas

# Latih dengan GD (full-batch)
model_gd = LinearRegressionGD(
    lr=lr_best,
    epochs=epochs,
    penalty=penalty_best,
    lam=lam_best,
    batch_size=None
).fit(X_train_bias, y_train)

# Latih dengan SGD (batch_size=1)
model_sgd = LinearRegressionGD(
    lr=lr_best,
    epochs=epochs,
    penalty=penalty_best,
    lam=lam_best,
    batch_size=1
).fit(X_train_bias, y_train)

# Latih dengan mini-batch SGD (batch_size=16)
model_mini_batch = LinearRegressionGD(
    lr=lr_best,
    epochs=epochs,
    penalty=penalty_best,
    lam=lam_best,
    batch_size=16
).fit(X_train_bias, y_train)

# 6.2 Visualisasi perbandingan konvergensi loss
plt.figure(figsize=(12, 6))
plt.plot(model_gd.history, label='GD (Full-batch)')
plt.plot(model_sgd.history, label='SGD (Batch=1)')
plt.plot(model_mini_batch.history, label='Mini-batch SGD (Batch=16)')
plt.title('Perbandingan Konvergensi Loss: GD vs SGD vs Mini-batch', fontsize=15)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss (MSE)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.savefig('gd_vs_sgd_convergence.png')
plt.show()

# 6.3 Analisis dengan smoothing untuk SGD (karena fluktuatif)
def smooth(y, box_pts):
    box = np.ones(box_pts) / box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth

plt.figure(figsize=(12, 6))
plt.plot(model_gd.history, label='GD (Full-batch)', alpha=0.9)
plt.plot(smooth(model_sgd.history, 20), label='SGD (Batch=1, Smoothed)', alpha=0.9)
plt.plot(smooth(model_mini_batch.history, 10), label='Mini-batch (Batch=16, Smoothed)', alpha=0.9)
plt.title('Perbandingan Konvergensi Loss dengan Smoothing', fontsize=15)
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss (MSE)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.savefig('smoothed_convergence_comparison.png')
plt.show()

# 6.4 Visualisasi efek regularisasi pada koefisien
# Latih model dengan regularisasi berbeda (tanpa, L1, L2)
models_reg = {}
for penalty in [None, 'l1', 'l2']:
    name = "No Regularization" if penalty is None else f"{penalty} Regularization"
    models_reg[name] = LinearRegressionGD(
        lr=0.1,
        epochs=500,
        penalty=penalty,
        lam=0.1 if penalty else 0.0,
        batch_size=None
    ).fit(X_train_bias, y_train)

# Plot perbandingan koefisien
coef_labels = ['Intercept'] + [f'Coef {i+1}' for i in range(X_train.shape[1])]
coef_data = []

for name, model in models_reg.items():
    coefs = model.theta
    coef_data.append(pd.Series(coefs, index=coef_labels, name=name))

coef_df = pd.DataFrame(coef_data).T

plt.figure(figsize=(12, 8))
coef_df.plot(kind='bar', figsize=(12, 8))
plt.title('Perbandingan Koefisien dengan Berbagai Tipe Regularisasi', fontsize=15)
plt.xlabel('Koefisien', fontsize=12)
plt.ylabel('Nilai', fontsize=12)
plt.legend(title='Regularisasi', fontsize=10)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('coefficient_comparison.png')
plt.show()

# 7. Analisis Hasil dan Penyajian Comprehensif
print("\n---- Analisis Hasil Akhir ----")

print("""
Analisis Hasil:
1. Perbandingan GD dan SGD:
   - GD (full-batch) lebih stabil tetapi memerlukan lebih banyak iterasi untuk konvergen
   - SGD (batch=1) lebih cepat per iterasi namun loss curve lebih fluktuatif (berisik)
   - Mini-batch SGD menawarkan keseimbangan antara kecepatan dan stabilitas

2. Pengaruh Learning Rate:
   - Learning rate kecil (0.01) mengakibatkan konvergensi lambat tapi stabil
   - Learning rate besar (0.5-1.0) mempercepat konvergensi tetapi berisiko osilasi atau divergensi
   - Learning rate optimal tergantung pada dataset dan kompleksitas model

3. Perbedaan L1 vs L2 Regularization:
   - L1 (Lasso) cenderung menghasilkan sparsity pada koefisien (beberapa koefisien mendekati nol)
   - L2 (Ridge) melakukan shrinkage pada semua koefisien secara merata tanpa sparsity
   - Pemilihan regularisasi tergantung pada kebutuhan: L1 untuk feature selection, L2 untuk stabilitas

4. Efek Batch Size pada SGD:
   - Batch kecil (1) menghasilkan update yang lebih noise tetapi lebih sering
   - Batch besar lebih stabil tetapi komputasi per iterasi lebih tinggi
   - Mini-batch (16, 32) menawarkan keseimbangan yang baik

5. Hyperparameter Terbaik:
   - Kombinasi terbaik ditemukan pada Penalty: {best_model_params['Penalty']},
     Learning Rate: {best_model_params['Learning Rate']}, Lambda: {best_model_params['Lambda']},
     Batch Size: {best_model_params['Batch Size']}
   - MSE terendah: {best_model_params['MSE']:.4f}
   - R² tertinggi: {best_model_params['R2']:.4f}
""")

# 8. Prediksi dengan Model Terbaik
# Latih model terbaik dengan parameter optimal
best_penalty = None if best_model_params['Penalty'] == 'None' else best_model_params['Penalty']
best_batch = None if best_model_params['Batch Size'] == 'Full' else best_model_params['Batch Size']

best_model = LinearRegressionGD(
    lr=best_model_params['Learning Rate'],
    epochs=best_model_params['Epochs'],
    penalty=best_penalty,
    lam=best_model_params['Lambda'],
    batch_size=best_batch
).fit(X_train_bias, y_train)

# Prediksi pada data test
y_pred = best_model.predict(X_test_bias)

# Visualisasi prediksi vs aktual
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.title('Nilai Aktual vs Prediksi (Model Terbaik)', fontsize=15)
plt.xlabel('Nilai Aktual', fontsize=12)
plt.ylabel('Nilai Prediksi', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.savefig('actual_vs_predicted.png')
plt.show()

# Hitung residual
residuals = y_test - y_pred

# Plot residual
plt.figure(figsize=(10, 6))
plt.scatter(y_pred, residuals, alpha=0.7)
plt.axhline(y=0, color='r', linestyle='--')
plt.title('Residual Plot (Model Terbaik)', fontsize=15)
plt.xlabel('Nilai Prediksi', fontsize=12)
plt.ylabel('Residual', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.savefig('residual_plot.png')
plt.show()

# Distribusi residual
plt.figure(figsize=(10, 6))
sns.histplot(residuals, kde=True)
plt.title('Distribusi Residual (Model Terbaik)', fontsize=15)
plt.xlabel('Residual', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.savefig('residual_distribution.png')
plt.show()

print("\n---- Kesimpulan ----")
print(f"""
Kesimpulan:
1. Model terbaik menghasilkan MSE = {best_model_params['MSE']:.4f} dan R² = {best_model_params['R2']:.4f}
2. Regularisasi {best_model_params['Penalty']} dengan lambda = {best_model_params['Lambda']} memberikan performa terbaik
3. Learning rate optimal adalah {best_model_params['Learning Rate']}
4. Batch size {best_model_params['Batch Size']} memberikan keseimbangan yang baik antara kecepatan dan stabilitas
5. Implementasi regresi linear dengan regularisasi berhasil dilakukan dengan GD dan SGD
""")

print("Implementasi selesai! Visualisasi dan hasil analisis telah disimpan.")
```
