```python
# ============================================================================
# Linear Regression dengan L1 (Lasso) dan L2 (Ridge) Regularization
# Implementasi GD dan SGD yang Dioptimasi dengan Visualisasi Lengkap
# ============================================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import time
from itertools import product
from tqdm import tqdm

# Set styling untuk visualisasi
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_context("notebook", font_scale=1.1)
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

# =============================================================================
# 1. PERSIAPAN DATA YANG LEBIH BAIK
# =============================================================================
# Load dataset
df = pd.read_csv('Medical Cost Personal.csv')
print("="*80)
print("DATASET OVERVIEW")
print("="*80)
print(f"Dataset Shape: {df.shape}")
print("\nPeek at data:")
print(df.head())

# Info dan ringkasan statistik dataset
print("\nInformasi Dataset:")
print(df.info())
print("\nRingkasan Statistik:")
print(df.describe())

# Cek missing values
print("\nJumlah Missing Values per Kolom:")
print(df.isnull().sum())

# =============================================================================
# 1.2 PREPROCESSING DATA YANG LEBIH BAIK
# =============================================================================
# Drop missing values (jika ada)
df.dropna(inplace=True)

# PERBAIKAN 1: Log transform pada target variable (charges) karena distribusinya skewed
# Ini penting untuk mengurangi MSE secara signifikan
df['charges_log'] = np.log1p(df['charges'])  # log1p = log(1+x) untuk menghindari log(0)

# One-hot encoding untuk variabel kategorikal
print("\nKolom sebelum encoding:")
print(df.columns.tolist())
df_encoded = pd.get_dummies(df, drop_first=True).astype(float)
print("\nKolom setelah encoding:")
print(df_encoded.columns.tolist())

# PERBAIKAN 2: Feature Engineering
# Tambahkan interaksi antara variabel penting
df_encoded['age_bmi'] = df_encoded['age'] * df_encoded['bmi']
df_encoded['smoker_yes_bmi'] = df_encoded['smoker_yes'] * df_encoded['bmi']
df_encoded['smoker_yes_age'] = df_encoded['smoker_yes'] * df_encoded['age']

# Simpan kolom asli untuk visualisasi
df_original = df.copy()

# Pisahkan fitur dan target
# PERBAIKAN 3: Gunakan target yang sudah di-log transform
y = df_encoded.pop('charges_log').values
y_original = df_encoded.pop('charges').values
X = df_encoded.values

# PERBAIKAN 4: Standardisasi fitur menggunakan RobustScaler untuk handle outliers lebih baik
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data menjadi train-test dengan stratifikasi (80% train, 20% test)
X_train, X_test, y_train, y_test, y_train_orig, y_test_orig = train_test_split(
    X_scaled, y, y_original, test_size=0.2, random_state=42
)

# Tambahkan bias (intercept)
def add_bias(X):
    return np.hstack([np.ones((X.shape[0], 1)), X])

X_train_bias = add_bias(X_train)
X_test_bias = add_bias(X_test)

print(f"\nDimensi Data Training setelah preprocessing: {X_train_bias.shape}")
print(f"Dimensi Data Testing setelah preprocessing: {X_test_bias.shape}")

# =============================================================================
# 2. EXPLORATORY DATA ANALYSIS (EDA)
# =============================================================================
print("\n" + "="*80)
print("EXPLORATORY DATA ANALYSIS")
print("="*80)

# 2.1 Distribusi target variable (charges) - Original dan Log-transformed
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Original distribution
sns.histplot(df_original['charges'], bins=30, kde=True, color='#3498db', ax=axes[0])
axes[0].set_title('Distribusi Biaya (Charges) - Original', fontsize=15, fontweight='bold')
axes[0].set_xlabel('Charges', fontsize=12)
axes[0].set_ylabel('Frekuensi', fontsize=12)
axes[0].grid(True, alpha=0.3, linestyle='--')

# Log-transformed distribution
sns.histplot(np.log1p(df_original['charges']), bins=30, kde=True, color='#e74c3c', ax=axes[1])
axes[1].set_title('Distribusi Biaya (Charges) - Log Transformed', fontsize=15, fontweight='bold')
axes[1].set_xlabel('Log(Charges+1)', fontsize=12)
axes[1].set_ylabel('Frekuensi', fontsize=12)
axes[1].grid(True, alpha=0.3, linestyle='--')

plt.tight_layout()
plt.savefig('charges_distribution_comparison.png', dpi=300, bbox_inches='tight')
plt.show()

# 2.2 Matriks korelasi fitur
plt.figure(figsize=(12, 10))
correlation_matrix = df_encoded.corr()
mask = np.triu(correlation_matrix)
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm',
            mask=mask, linewidths=0.5, annot_kws={"size": 10})
plt.title('Matriks Korelasi Fitur (Termasuk Fitur Baru)', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('correlation_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

# 2.3 Scatter plot antara fitur numerik dan target
numeric_cols = ['age', 'bmi', 'children']
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
colors_palette = ['#3498db', '#2ecc71', '#e74c3c']

for i, col in enumerate(numeric_cols):
    sns.scatterplot(data=df_original, x=col, y='charges', ax=axes[i], color=colors_palette[i], alpha=0.7, edgecolor='k', s=70)
    axes[i].set_title(f'Charges vs {col.capitalize()}', fontsize=13, fontweight='bold')
    axes[i].set_xlabel(col.capitalize(), fontsize=11)
    axes[i].set_ylabel('Charges', fontsize=11)
    axes[i].grid(True, alpha=0.3, linestyle='--')

plt.tight_layout()
plt.savefig('feature_target_relationships.png', dpi=300, bbox_inches='tight')
plt.show()

# 2.4 Box plot untuk fitur kategorikal
categorical_cols = ['sex', 'smoker', 'region']
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
colors_palette = ['#9b59b6', '#f39c12', '#1abc9c']

for i, col in enumerate(categorical_cols):
    sns.boxplot(data=df_original, x=col, y='charges', ax=axes[i], palette=[colors_palette[i], '#bdc3c7'])
    axes[i].set_title(f'Charges by {col.capitalize()}', fontsize=13, fontweight='bold')
    axes[i].set_xlabel(col.capitalize(), fontsize=11)
    axes[i].set_ylabel('Charges', fontsize=11)
    axes[i].grid(True, alpha=0.3, linestyle='--')

plt.tight_layout()
plt.savefig('categorical_features_boxplots.png', dpi=300, bbox_inches='tight')
plt.show()

# 2.5 Visualisasi khusus untuk fitur interaksi yang baru ditambahkan
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
interaction_cols = ['age_bmi', 'smoker_yes_bmi', 'smoker_yes_age']
df_vis = df_encoded.copy()
df_vis['charges'] = y_original

for i, col in enumerate(interaction_cols):
    sns.scatterplot(data=df_vis, x=col, y='charges', ax=axes[i], color=colors_palette[i], alpha=0.7, edgecolor='k', s=70)
    axes[i].set_title(f'Charges vs {col}', fontsize=13, fontweight='bold')
    axes[i].set_xlabel(col, fontsize=11)
    axes[i].set_ylabel('Charges', fontsize=11)
    axes[i].grid(True, alpha=0.3, linestyle='--')

plt.tight_layout()
plt.savefig('interaction_features.png', dpi=300, bbox_inches='tight')
plt.show()

# =============================================================================
# 3. DEFINISI MODEL REGRESI LINEAR DENGAN REGULARISASI
# =============================================================================
class LinearRegressionGD:
    """
    Implementasi Linear Regression dengan Gradient Descent dan regularisasi.

    PERBAIKAN: Tambahkan early stopping dan learning rate scheduling
    """

    def __init__(self, lr=0.01, epochs=1000, penalty=None, lam=0.0, batch_size=None,
                 lr_decay=False, early_stopping=True, patience=50):
        self.lr = lr
        self.epochs = epochs
        self.penalty = penalty  # None, 'l1', 'l2'
        self.lam = lam
        self.batch_size = batch_size
        self.theta = None
        self.history = []  # untuk menyimpan loss per epoch
        self.theta_history = []  # untuk menyimpan koefisien per epoch
        self.lr_decay = lr_decay  # PERBAIKAN: tambahkan learning rate decay
        self.early_stopping = early_stopping  # PERBAIKAN: early stopping
        self.patience = patience  # epoch untuk menunggu sebelum early stopping
        self.best_loss = float('inf')
        self.no_improvement_count = 0

    def _compute_gradient(self, X_batch, y_batch):
        """Menghitung gradient dari loss function dengan/tanpa regularisasi"""
        m = X_batch.shape[0]
        preds = X_batch.dot(self.theta)
        error = preds - y_batch

        # Gradient dari MSE
        grad = X_batch.T.dot(error) / m

        # Tambahkan gradient dari penalty regularisasi
        if self.penalty == 'l2':  # Ridge: L2 penalty
            # Tidak menerapkan regularisasi pada bias (theta[0])
            grad[1:] += (self.lam / m) * self.theta[1:]
        elif self.penalty == 'l1':  # Lasso: L1 penalty
            # Tidak menerapkan regularisasi pada bias (theta[0])
            grad[1:] += (self.lam / m) * np.sign(self.theta[1:])

        return grad

    def _compute_loss(self, X, y):
        """Menghitung loss (MSE + regularisasi)"""
        m = X.shape[0]
        preds = X.dot(self.theta)
        mse = np.mean((preds - y) ** 2)

        # Tambahkan penalty regularisasi ke loss
        if self.penalty == 'l2':  # Ridge: L2 penalty
            mse += (self.lam / (2 * m)) * np.sum(self.theta[1:] ** 2)
        elif self.penalty == 'l1':  # Lasso: L1 penalty
            mse += (self.lam / m) * np.sum(np.abs(self.theta[1:]))

        return mse

    def fit(self, X, y):
        """Fit model dengan Gradient Descent atau Stochastic Gradient Descent"""
        n_samples, n_features = X.shape

        # PERBAIKAN: Inisialisasi dengan nilai kecil acak untuk mencegah local minima
        self.theta = np.random.randn(n_features) * 0.01

        # Tentukan batch size
        bs = self.batch_size if self.batch_size else n_samples

        # Simpan theta terbaik
        best_theta = None

        # Learning rate awal
        current_lr = self.lr

        # Training loop
        for epoch in range(self.epochs):
            # PERBAIKAN: Learning rate decay
            if self.lr_decay:
                current_lr = self.lr / (1 + 0.01 * epoch)

            # Shuffle data untuk SGD
            if bs < n_samples:
                indices = np.random.permutation(n_samples)
                X_shuffled, y_shuffled = X[indices], y[indices]
            else:
                X_shuffled, y_shuffled = X, y

            # Mini-batch updates
            for start_idx in range(0, n_samples, bs):
                end_idx = min(start_idx + bs, n_samples)
                X_batch = X_shuffled[start_idx:end_idx]
                y_batch = y_shuffled[start_idx:end_idx]

                # Hitung gradient dan update parameters
                grad = self._compute_gradient(X_batch, y_batch)
                self.theta -= current_lr * grad

                # Early stopping jika weights tidak finite
                if not np.isfinite(self.theta).all():
                    print(f"Warning: Non-finite weights detected at epoch {epoch}. Training stopped.")
                    return self

            # Hitung dan simpan loss tiap epoch
            loss = self._compute_loss(X, y)
            self.history.append(loss)
            self.theta_history.append(self.theta.copy())

            # Early stopping jika loss tidak finite
            if not np.isfinite(loss):
                print(f"Warning: Non-finite loss detected at epoch {epoch}. Training stopped.")
                return self

            # PERBAIKAN: Early stopping check
            if self.early_stopping:
                if loss < self.best_loss:
                    self.best_loss = loss
                    best_theta = self.theta.copy()
                    self.no_improvement_count = 0
                else:
                    self.no_improvement_count += 1

                if self.no_improvement_count >= self.patience:
                    print(f"Early stopping at epoch {epoch}, best loss: {self.best_loss:.6f}")
                    self.theta = best_theta  # Restore best weights
                    break

            # Print progress setiap 100 epoch atau di epoch terakhir
            if epoch % 100 == 0 or epoch == self.epochs - 1:
                print(f"Epoch {epoch}/{self.epochs}, Loss: {loss:.6f}, LR: {current_lr:.6f}")

        # Jika early stopping tidak terpicu, gunakan theta terbaik
        if self.early_stopping and best_theta is not None:
            self.theta = best_theta

        return self

    def predict(self, X):
        """Prediksi dengan model yang sudah dilatih"""
        return X.dot(self.theta)

    def get_params(self):
        """Return model parameters"""
        return {'intercept': self.theta[0], 'coefficients': self.theta[1:]}

# =============================================================================
# 4. PENGATURAN HYPERPARAMETER DAN TUNING
# =============================================================================
print("\n" + "="*80)
print("HYPERPARAMETER TUNING")
print("="*80)

# PERBAIKAN: Range hyperparameter yang lebih baik
hyperparams = {
    'lr': [0.001, 0.01, 0.1, 0.5],
    'epochs': [500, 1000, 2000],
    'batch_size': [8, 16, 32, None],
    'lam': [0.0001, 0.001, 0.01, 0.1]
}

# Buat semua kombinasi hyperparameter
combinations = []
for penalty in [None, 'l1', 'l2']:
    # Hanya menggunakan lambda > 0 jika ada regularisasi
    lam_list = hyperparams['lam'] if penalty else [0.0]

    for lr, epochs, bs, lam in product(
        hyperparams['lr'],
        hyperparams['epochs'],
        hyperparams['batch_size'],
        lam_list
    ):
        combinations.append({
            'penalty': penalty,
            'lr': lr,
            'epochs': epochs,
            'batch_size': bs,
            'lam': lam,
            'lr_decay': True,  # PERBAIKAN: selalu gunakan LR decay
            'early_stopping': True  # PERBAIKAN: selalu gunakan early stopping
        })

print(f"Total kombinasi hyperparameter: {len(combinations)}")

# Proses tuning dengan semua kombinasi
results = []
for params in tqdm(combinations, desc='Tuning Models'):
    # Set parameter untuk model ini
    model_params = params.copy()

    try:
        # Catat waktu mulai
        t0 = time.time()

        # Inisialisasi dan latih model
        model = LinearRegressionGD(**model_params)
        model.fit(X_train_bias, y_train)

        # Prediksi pada data test
        y_pred = model.predict(X_test_bias)

        # Evaluasi model
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)

        # Catat waktu selesai
        t1 = time.time()

        # Simpan hasil
        results.append({
            'Penalty': params['penalty'] if params['penalty'] else 'None',
            'Learning Rate': params['lr'],
            'Epochs': params['epochs'],
            'Batch Size': params['batch_size'] if params['batch_size'] else 'Full',
            'Lambda': params['lam'],
            'MSE': mse,
            'R2': r2,
            'Training Time (s)': t1 - t0
        })
    except Exception as e:
        print(f"Error with params {params}: {e}")
        results.append({
            'Penalty': params['penalty'] if params['penalty'] else 'None',
            'Learning Rate': params['lr'],
            'Epochs': params['epochs'],
            'Batch Size': params['batch_size'] if params['batch_size'] else 'Full',
            'Lambda': params['lam'],
            'MSE': np.nan,
            'R2': np.nan,
            'Training Time (s)': np.nan
        })

# =============================================================================
# 5. EVALUASI MODEL
# =============================================================================
print("\n" + "="*80)
print("EVALUASI MODEL")
print("="*80)

# Buat DataFrame hasil dan hapus baris dengan nilai NaN
results_df = pd.DataFrame(results).dropna()

# Urutkan berdasarkan MSE (dari terendah ke tertinggi)
sorted_results = results_df.sort_values('MSE').reset_index(drop=True)

# Cetak 10 model terbaik
print("\nTop 10 Model Berdasarkan MSE Terendah:")
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: '%.4f' % x)
print(sorted_results.head(10))

# 5.1 Pilih model terbaik berdasarkan MSE untuk analisis lebih lanjut
best_model_params = sorted_results.iloc[0].to_dict()
print("\nParameter Model Terbaik:")
for key, value in {k: best_model_params[k] for k in
                  ['Penalty', 'Learning Rate', 'Epochs', 'Batch Size', 'Lambda']}.items():
    print(f"{key}: {value}")

# 5.2 Visualisasi perbandingan performa model berdasarkan tipe regularisasi
plt.figure(figsize=(12, 6))
sns.boxplot(x='Penalty', y='MSE', data=results_df, palette=['#3498db', '#e74c3c', '#2ecc71'])
plt.title('Perbandingan MSE Berdasarkan Tipe Regularisasi', fontsize=15, fontweight='bold')
plt.xlabel('Tipe Regularisasi', fontsize=12)
plt.ylabel('Mean Squared Error (MSE)', fontsize=12)
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()
plt.savefig('regularization_comparison_mse.png', dpi=300, bbox_inches='tight')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='Penalty', y='R2', data=results_df, palette=['#3498db', '#e74c3c', '#2ecc71'])
plt.title('Perbandingan R Berdasarkan Tipe Regularisasi', fontsize=15, fontweight='bold')
plt.xlabel('Tipe Regularisasi', fontsize=12)
plt.ylabel('R Score', fontsize=12)
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()
plt.savefig('regularization_comparison_r2.png', dpi=300, bbox_inches='tight')
plt.show()

# 5.3 Visualisasi perbandingan performa model berdasarkan batch size
plt.figure(figsize=(12, 6))
sns.boxplot(x='Batch Size', y='MSE', data=results_df, palette=['#f39c12', '#9b59b6', '#1abc9c', '#34495e'])
plt.title('Perbandingan MSE Berdasarkan Ukuran Batch', fontsize=15, fontweight='bold')
plt.xlabel('Ukuran Batch', fontsize=12)
plt.ylabel('Mean Squared Error (MSE)', fontsize=12)
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()
plt.savefig('batch_size_comparison_mse.png', dpi=300, bbox_inches='tight')
plt.show()

# 5.4 Visualisasi pengaruh learning rate
plt.figure(figsize=(12, 6))
sns.boxplot(x='Learning Rate', y='MSE', data=results_df, palette=['#1abc9c', '#3498db', '#9b59b6', '#e74c3c'])
plt.title('Perbandingan MSE Berdasarkan Learning Rate', fontsize=15, fontweight='bold')
plt.xlabel('Learning Rate', fontsize=12)
plt.ylabel('Mean Squared Error (MSE)', fontsize=12)
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()
plt.savefig('learning_rate_comparison_mse.png', dpi=300, bbox_inches='tight')
plt.show()

# 5.5 Visualisasi pengaruh lambda (strength regularisasi)
plt.figure(figsize=(12, 6))
sns.boxplot(x='Lambda', y='MSE', data=results_df[results_df['Penalty'] != 'None'],
            palette=['#2ecc71', '#3498db', '#e74c3c', '#9b59b6'])
plt.title('Perbandingan MSE Berdasarkan Lambda (Regularization Strength)', fontsize=15, fontweight='bold')
plt.xlabel('Lambda', fontsize=12)
plt.ylabel('Mean Squared Error (MSE)', fontsize=12)
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()
plt.savefig('lambda_comparison_mse.png', dpi=300, bbox_inches='tight')
plt.show()

# =============================================================================
# 6. PERBANDINGAN GD VS SGD
# =============================================================================
print("\n" + "="*80)
print("PERBANDINGAN GD VS SGD")
print("="*80)

# Ekstrak parameter model terbaik
best_penalty = best_model_params['Penalty'] if best_model_params['Penalty'] != 'None' else None
best_lr = best_model_params['Learning Rate']
best_epochs = 500  # menggunakan 500 epochs untuk visualisasi yang jelas
best_lambda = best_model_params['Lambda']

# 6.1 Latih model GD (full-batch) dan SGD dengan parameter terbaik
# Latih dengan GD (full-batch)
print("Training model GD (full-batch)...")
model_gd = LinearRegressionGD(
    lr=best_lr,
    epochs=best_epochs,
    penalty=best_penalty,
    lam=best_lambda,
    batch_size=None,
    lr_decay=True,
    early_stopping=False  # Matikan early stopping untuk visualisasi lengkap
).fit(X_train_bias, y_train)

# Latih dengan SGD (batch_size=1)
print("Training model SGD (batch_size=1)...")
model_sgd = LinearRegressionGD(
    lr=best_lr,
    epochs=best_epochs,
    penalty=best_penalty,
    lam=best_lambda,
    batch_size=1,
    lr_decay=True,
    early_stopping=False  # Matikan early stopping untuk visualisasi lengkap
).fit(X_train_bias, y_train)

# Latih dengan mini-batch SGD (batch_size=16)
print("Training model mini-batch SGD (batch_size=16)...")
model_mini_batch = LinearRegressionGD(
    lr=best_lr,
    epochs=best_epochs,
    penalty=best_penalty,
    lam=best_lambda,
    batch_size=16,
    lr_decay=True,
    early_stopping=False  # Matikan early stopping untuk visualisasi lengkap
).fit(X_train_bias, y_train)

# 6.2 Visualisasi perbandingan konvergensi loss
plt.figure(figsize=(12, 6))
plt.plot(model_gd.history, label='GD (Full-batch)', color='#3498db', linewidth=2)
plt.plot(model_sgd.history, label='SGD (Batch=1)', color='#e74c3c', linewidth=2, alpha=0.6)
plt.plot(model_mini_batch.history, label='Mini-batch SGD (Batch=16)', color='#2ecc71', linewidth=2)
plt.title('Perbandingan Konvergensi Loss: GD vs SGD vs Mini-batch', fontsize=15, fontweight='bold')
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss (MSE)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()
plt.savefig('gd_vs_sgd_convergence.png', dpi=300, bbox_inches='tight')
plt.show()

# 6.3 Analisis dengan smoothing untuk SGD (karena fluktuatif)
def smooth(y, box_pts):
    box = np.ones(box_pts) / box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth

plt.figure(figsize=(12, 6))
plt.plot(model_gd.history, label='GD (Full-batch)', color='#3498db', linewidth=2)
plt.plot(smooth(model_sgd.history, 20), label='SGD (Batch=1, Smoothed)', color='#e74c3c', linewidth=2)
plt.plot(smooth(model_mini_batch.history, 10), label='Mini-batch (Batch=16, Smoothed)', color='#2ecc71', linewidth=2)
plt.title('Perbandingan Konvergensi Loss dengan Smoothing', fontsize=15, fontweight='bold')
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss (MSE)', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()
plt.savefig('smoothed_convergence_comparison.png', dpi=300, bbox_inches='tight')
plt.show()

# 6.4 Visualisasi efek regularisasi pada koefisien
# Latih model dengan regularisasi berbeda (tanpa, L1, L2)
models_reg = {}
for penalty in [None, 'l1', 'l2']:
    name = "No Regularization" if penalty is None else f"{penalty} Regularization"
    print(f"Training model dengan {name}...")
    models_reg[name] = LinearRegressionGD(
        lr=0.1,
        epochs=500,
        penalty=penalty,
        lam=0.1 if penalty else 0.0,
        batch_size=None,
        lr_decay=True,
        early_stopping=False
    ).fit(X_train_bias, y_train)

# Plot perbandingan koefisien
coef_labels = ['Intercept'] + [f'Coef {i+1}' for i in range(X_train.shape[1])]
coef_data = []

for name, model in models_reg.items():
    coefs = model.theta
    coef_data.append(pd.Series(coefs, index=coef_labels, name=name))

coef_df = pd.DataFrame(coef_data).T

plt.figure(figsize=(12, 8))
colors = ['#3498db', '#e74c3c', '#2ecc71']
coef_df.plot(kind='bar', figsize=(12, 8), color=colors)
plt.title('Perbandingan Koefisien dengan Berbagai Tipe Regularisasi', fontsize=15, fontweight='bold')
plt.xlabel('Koefisien', fontsize=12)
plt.ylabel('Nilai', fontsize=12)
plt.legend(title='Regularisasi', fontsize=10)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3, linestyle='--')
plt
```
